```python
#importar librerias
import numpy as np
from numpy import sin, cos, tan, arctan, pi, matmul,arange
import matplotlib.pyplot as plt
import random as rd
from random import randint, random, uniform
```

```python
#datos
alfa=1 #factor aprendizaje
p=np.array([   [0,0,1,1],[0,1,0,1]   ])
t=np.array([1,1,1,0])
#W=np.array([0.5,1.5])
W=np.random.rand(2)
#b=1.5
b=np.random.rand(1)
```

```python
#funcion
def escalon1(x):
    if x<0:
        y=0
    else:
        y=1
    return y 
#r1=escalon(0)
#print(r1)
```

```python
#ciclos de Aprendizaje
errorT = 1
jj = 0
while errorT != 0:# Ciclos de aprendizaje (iterar)    
    errorT = 0    
    for k in range(4):
        pp = p[:, k]
        aa = matmul(W, pp) + b
        a = escalon1(aa)
        error = t[k] - a
        W = W + alfa * error * pp.transpose()
        b = b + alfa * error * 1
        errorT = errorT + abs(error)        
        # print(pp)
        # print(aa, "  ", a, "  ", error)
        # print(W, "  ", b)       
     
    #end k
    print("jj=", jj, "  ", errorT)
    print(W, "  ", b)  
    jj += 1 
    #if errorT == 0:
        #break
#end while
```
